### 1. Python 기본기 다시 잡아보기!

- 변수(Variable): "변할수 있는 수(정보)", 정보들을 담아둘 수 있는 공간

```python
my_name = "" # 변수의 이름은 자유롭게 설정가능!
my_age = 0

print("이름: " + my_name + "\n나이: " + str(my_age))
```

- 함수(Function): 입력을 받아 어떤 일을 수행하고 출력하는 장치, 호출한 곳에 한 일의 결과를 반환함

```python
def my_func1(input): # 함수의 이름은 자유롭게 설정가능!
    output = input + 1

    return output

my_number = 10

print("my_number 더하기 1 은: " + str(my_func1(10)))

# print() 는 터미널에 출력하는 함수
# str() 은 숫자 등을 문자열로 바꿔주는 함수
```

- 클래스(Class): 여러 변수나 함수들을 한 자리에 묶어, 하나의 기능으로 활용할 수 있도록 만들어 주는 기능의 모임

```python
# 클래스의 이름은 자유롭게 설정가능!
class MyClass:
    # 초기생성자(Initializer), 클래스로 객체를 만들면 맨 처음 무엇을 할지 정해주는 함수. 앞 뒤로 언더바 2개씩 필수!
    def __init__(self):
        self.my_var1 = 0;

    def my_func2(self, input):
        output = self.my_var1 + input

        return output


# 클래스로 만들어진 변수는 객체(Object)라고 부름
my_object = MyClass()
# . 을 이용해 객체 안의 내용물에 진입가능
my_object.my_var1 = 10
my_number = 20

print("my_var1  : " + str(my_object.my_var1))
print("my_number: " + str(my_number))

# 클래스 내에 정의된 함수들도 . 으로 진입 가능!
print("두개 합치면: " + str(my_object.my_func2(my_number)))
```

### 2. Python 의 자료형들 알아보기!

- **문자(String) 형식**: str <br>
주로 따옴표 "", '' 안에 들어가는 것들. <br>

```python
my_string1 = "Hello World!"
my_string2 = 'Hello World!'

print(type(my_string1))
print(type(my_string2))
```

- **숫자(Numeric) 형식**: int, float, complex <br>
정수, 소수점 있는 수나 허수등. <br>

```python
my_int = 10
my_float = 12.345
my_complex = 1.23j # j 에 주목!

print(type(my_int))
print(type(my_float))
print(type(my_complex))
```

- **열거(Sequence) 형식**: list, tuple, range <br>
리스트, 튜플과 같이 정보들이 나열된 것들.<br>

```python
my_list = ["Apple", "Banana", "Cherry"]
my_tuple = ("Apple", "Banana", "Cherry")
my_range = range(10)

print(type(my_list))
print(type(my_tuple))
print(type(my_range))
```

- **대입(Mapping) 형식**: dict <br>
콜론(: 기호)를 이용해 데이터의 짝을 이루는 것들. 파이썬에선 데이터를 받아오거나, JSON 파일 등의 데이터 파일을 다룰때 굉장히 자주 사용됨.

```python
my_dict = {"name" : "John", "Age" : 30}

print(type(my_dict))
```

- **집합(Set) 형식**: set, frozenset <br>
집합, 즉 정보들을 그룹 지은것. frozenset은 내용물을 변경 할수 없게 잠긴 set 형식.<br>

```python
my_set = {"Apple", "Banana", "Cherry"}
my_frozenset = frozenset({"Apple", "Banana", "Cherry"})

print(type(my_set))
print(type(my_frozenset))
```

- **참거짓(Boolean) 형식**: bool <br>
조건문이 참인지 거짓인지 판단할 때 쓰이는 자료형. True(참), False(거짓) 으로 나뉘며, if 등 조건문을 사용하는 식에 응용됨. <br>

```python
my_bool = True

print(type(my_bool))
```

- **이진(Binary) 형식**: bytes, bytearray, memoryview <br>
컴퓨터의 가장 기본이 되는 이진수(0과 1) 형식을 나타내는 자료형들. <br>

```python
my_bytes = b"Hello"
my_bytearray = bytearray(5)
my_memoryview = memoryview(bytes(5))

print(type(my_bytes))
print(type(my_bytearray))
print(type(my_memoryview))
```

- **형식 없음(None)**: NoneType<br>
현재로서 아무것도 가지지 않은 상태를 표현하기 위한 자료형.<br>
다른 언어에선 null 과 같은 형태로 볼 수 있음.<br>

```python
my_none = None

print(type(my_none))
```

### 3. Jupyter 노트북의 애니메이션 감상해보기!

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# 회전 행렬
def rotation_matrix(theta):
    return np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta),  np.cos(theta)]
    ])

# 그래프 설정
fig, ax = plt.subplots(figsize=(6, 6))
vector = np.array([1, 0])

# 빈 quiver 객체 생성
original_quiver = ax.quiver(0, 0, *vector, color='blue', scale=1, scale_units='xy', angles='xy', label='Original', width=0.01)
rotated_quiver = ax.quiver(0, 0, 0, 0, color='red', scale=1, scale_units='xy', angles='xy', label='Rotated', width=0.01)

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend()

# 각도 배열
angles = np.linspace(0, 2*np.pi, 100)

# 애니메이션을 돌릴 때 각 프레임마다 호출되는 함수
def update(frame):
    R = rotation_matrix(angles[frame])
    rotated = R @ vector
    rotated_quiver.set_UVC(rotated[0], rotated[1])
    
    return rotated_quiver,

anim = FuncAnimation(
    fig,
    update,
    frames=len(angles),
    interval=50,
    blit=True,
    repeat=True
)

plt.close()  # 정적 이미지 방지
HTML(anim.to_jshtml())  # 인터랙티브 HTML 플레이어
```

### 4. 정렬 알고리즘 알아보기!

```python
def bubble_sort_steps(arr):
    """
    버블 정렬의 모든 단계를 기록
    
    Returns:
        steps: [(data, comparing_indices, sorted_indices), ...]
               각 단계마다 (배열 상태, 비교중인 인덱스들, 정렬완료 인덱스들)
    """
    data = arr.copy()
    n = len(data)
    steps = []
    
    steps.append((data.copy(), [], set()))
    
    for i in range(n): # 이게 뭘까?
        for j in range(0, n - i - 1): # 왜 1을 빼고 있을까?
            steps.append((data.copy(), [j, j+1], set(range(n-i, n))))
            
            if data[j + 1] < data[j]:
                data[j], data[j + 1] = data[j + 1], data[j]
                
                # 윗 줄은 파이썬 내부에서 튜플을 만들어 두개를 동시에 대입하기 때문에 문제없음.
                # temp = data[j]
                # data[j] = data[j + 1]
                # data[j + 1] = temp

                steps.append((data.copy(), [j, j+1], set(range(n-i, n))))
        
        steps.append((data.copy(), [], set(range(n-i-1, n))))

    steps.append((data.copy(), [], set(range(n))))
    
    return steps


# 랜덤 데이터 생성
np.random.seed(123)
data = np.random.randint(1, 100, size=25)

# 정렬 과정 기록
steps = bubble_sort_steps(data)

# 그래프 설정
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(-0.5, len(data) - 0.5)
ax.set_ylim(0, max(data) * 1.1)
ax.set_xlabel("index", fontsize=12)
ax.set_ylabel("value", fontsize=12)
ax.set_title("Bubble Sort", fontsize=14, fontweight='bold')

# 막대 그래프 초기화
bars = ax.bar(
    range(len(data)),
    data,
    color='skyblue',
    edgecolor='black',
    linewidth=0.5
)

# 죄측 상단 텍스트 표시용
info_text = ax.text(
    0.02,
    0.95,
    '',
    transform=ax.transAxes,
    verticalalignment='top',
    fontsize=11,
    bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5)
)

# 애니메이션을 돌릴 때 각 프레임마다 호출되는 함수
def update(frame):
    current_data, comparing, sorted_indices = steps[frame]
    
    for i, (bar, value) in enumerate(zip(bars, current_data)):
        bar.set_height(value)
        
        if i in sorted_indices:
            bar.set_color('limegreen')
        elif i in comparing:
            bar.set_color('crimson')
        else:
            bar.set_color('skyblue')
    
    info_text.set_text(
        f"Step: {frame}/{len(steps)-1}\n"
        f"Comparing: {comparing if comparing else 'None'}\n"
        f"Finished: {len(sorted_indices)}"
    )
    
    return bars


anim = FuncAnimation(
    fig, 
    update, 
    frames=len(steps), 
    interval=10, # 애니메이션 속도 조정
    blit=False,  # 전체 다시 그리기 (텍스트 업데이트 위해 필요)
    repeat=False # 반복 재생 끄기
)

plt.close()
HTML(anim.to_jshtml())  # Jupyter에서 인터랙티브 플레이어로 표시
```
