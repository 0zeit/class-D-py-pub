### 2. 우리는 정보(데이터)를 어떻게 저장할까? 📦

그런데 한 번 생각해볼까요? 카카오톡에서 친구랑 대화할 때, 내 이름이나 친구 이름, 그리고 채팅 내용은 어디에 저장될까요? <br>
폰이나 컴퓨터는 이런 정보를 어떻게 기억하는 걸까요? <br>
<br>
프로그래밍은 이걸 "변수"로 해결합니다. "변할 수 있는 수(정보)"의 줄임말이죠. <br>
**변수는 쉽게 말하면 "뭐든 넣을 수 있는 이름이 붙은 상자" 와 같아요.** <br>
변수라는 개념은 생소하게 다가올 수도 있어요. 일단 변수 없이 간단한 게임을 만들면 어떤지 한번 볼까요? <br>

```python
print("철수의 체력:", 100)
print("공격을 받았어요!")     # 공격을 얼마나 받았지?
print("철수의 체력:", 85)    # 어? 100이었는데 85로 바뀌었네?
print("회복 물약을 먹었어요!") # 얼마나 회복한거지?
print("철수의 체력:", 100)   # 숫자를 매번 번거롭게 써야하나...?
```

위 코드의 예제에선, 철수가 공격을 받고 회복을 하는걸 보여주는데, 여기서 생각해 볼 점들은: <br>
1. 철수의 체력이 처음부터 100?
2. 공격을 얼마나 받은거지?
3. 회복물약으로 얼마나 회복한거지?

이와 같은 부분들이 의문점으로 남게되죠. <br>
만약 철수의 체력이 게임을 하면서 100번 바뀌게 된다면, 똑같이 100번 숫자를 써내려가야 하는 번거로움이 생겨요. <br>
파이썬(Python)을 배우면서, 우리는 "변수"를 활용해 어떤 일이 일어나고 있는가 더욱 확실하게 만들 수 있어요. <br>

```python
name = "철수" # name, 이름
health = 100 # health, 체력

print(name + "의 체력:", health)
```

위 코드에서 게임의 주인공 이름을 정해주고, 체력도 정해줬어요. <br>
`name` 과 `health` 가 여기서 "변수" 역할을 하고 있고, 이 변수들은 기호 = 을 이용해 만들 수 있어요. <br>
기호 = 입장에서, 오른쪽이 변수라는 상자에 넣고싶은 "정보"이고, 왼쪽이 정보를 담을 상자가 되는거에요. <br>

```python
# 상자에 이름표 붙이고 물건 넣기
name = "철수"   # "name" 상자에 "철수" 저장
age = 15       # "age" 상자에 15 저장
score = 9500   # "score" 상자에 9500 저장
```
